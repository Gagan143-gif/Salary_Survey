```python
# Importing required libraries.
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline
sns.set(color_codes=True)

from google.colab import files
uploaded = files.upload()
```

```python


# Loading the data into the data frame.

df = pd.read_csv("Salary_Survey.csv")
# To display the top 5 rows
df.head()
```

```python
# To display the bottom 5 rows
df.tail()
```

```python
# Checking the data type
df.dtypes
```

```python
# Data Cleaning
# Dropping irrelevant columns
df = df.drop(['Unnamed: 17','Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22','Timestamp'], axis=1)
df.head()
```

```python
# Renaming the column names
df = df.rename(columns={"How old are you?": "Age", "What industry do you work in?": "Work field","If your job title needs additional context, please clarify here:": "Role Context", "What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)": "Equivalent full-time annual earnings (40 hrs/week, 52 weeks)","How much additional monetary compensation do you get, if any (for example, bonuses or overtime in an average year)? Please only include monetary compensation here, not the value of benefits.": "Additional monetary compensation (e.g., bonuses, overtime):", "Please indicate the currency": "Currency", "What country do you work in?": "Country", "How many years of professional work experience do you have overall?": "Professional Experience","What is your highest level of education completed?": "Highest level of education completed","What is your gender?": "Gender"})

df.head()
```

```python
# Total number of rows and columns
df.shape
```

```python
# Rows containing duplicate data
duplicate_rows_df = df[df.duplicated()]
# Removing duplicate rows
df.drop_duplicates(inplace=True)
print("number of duplicate rows:", duplicate_rows_df.shape)
```

```python
# Used to count the number of rows before removing the data
df.count()
```

```python
# Finding the null values.
print(df.isnull().sum())
```

```python
# Dropping the missing values.
df = df.dropna()
df.count()
```

```python
# After dropping the values
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
# Detecting Outliers
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['Equivalent full-time annual earnings (40 hrs/week, 52 weeks)'])
plt.title("Salary Distribution by Industry")
plt.xlabel("Industry")
plt.ylabel("Annual Salary")
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(12,8))
sns.boxplot(x=df['If you\'re in the U.S., what state do you work in?'])
plt.title("Salary Distribution by State")
plt.xlabel("State")
plt.ylabel("Annual Salary")
plt.xticks(rotation=45,fontsize=12)  # Rotate x-axis labels 45 degrees
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd  # Import pandas for data manipulation

# Assuming df is your DataFrame
# Convert 'Age' column to numeric, handling errors by coercing to NaN
df['Age_Numeric'] = pd.to_numeric(df['Age'], errors='coerce')

# Filter out NaN values to plot only numeric ages
data = df['Age_Numeric'].dropna()

# Plotting the histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.title("Age Distribution of Respondents")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plotting a Histogram
df.Timestamp.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Equivalent full-time annual earnings (40 hrs/week, 52 weeks)")
plt.ylabel('Age')
plt.xlabel('Additional monetary compensation (e.g., bonuses, overtime):')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title("Missing Data Heatmap")
plt.show()
```

```python
# Finding the relations between the variables.
plt.figure(figsize=(20,10))
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
c = numeric_df.corr()

# Now use 'c' in the heatmap
sns.heatmap(c,cmap="BrBG",annot=True)
plt.show() # It's helpful to add plt.show() to display the plot
```
